## K means

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
in_df = pd.read_csv(
    r'/Users/venkatesh/Workspace/Udemy/ML/Machine Learning A-Z Template Folder/Part 4 - Clustering/Section 24 - K-Means Clustering/K_Means/Mall_Customers.csv'
)
```

```python
x = in_df.iloc[:,[3,4]].values
```

```python
# Using Elbow method
from sklearn.cluster import KMeans

wcss = []

for i in range(1,11):
    kmeans = KMeans(
        n_clusters=i, init='k-means++', max_iter=300, random_state=0, n_init=10
    )
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
```

```python
kmeans = KMeans(
    n_clusters=5, init='k-means++', max_iter=300, random_state=0, n_init=10
)
pred_kmeans = kmeans.fit_predict(x)
```

```python
# Visualization
plt.scatter(x[pred_kmeans==0,0], x[pred_kmeans==0,1], s=100, c='red', label='cluster 1')
plt.scatter(x[pred_kmeans==1,0], x[pred_kmeans==1,1], s=100, c='blue', label='cluster 2')
plt.scatter(x[pred_kmeans==2,0], x[pred_kmeans==2,1], s=100, c='green', label='cluster 3')
plt.scatter(x[pred_kmeans==3,0], x[pred_kmeans==3,1], s=100, c='cyan', label='cluster 4')
plt.scatter(x[pred_kmeans==4,0], x[pred_kmeans==4,1], s=100, c='magenta', label='cluster 5')
plt.scatter(
    kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=300, c='yellow', label='centroids'
)
plt.xlabel("Annual Incom K$")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```

## Hierarchical Clustering

```python
in_df = pd.read_csv(
    r'/Users/venkatesh/Workspace/Udemy/ML/Machine Learning A-Z Template Folder/Part 4 - Clustering/Section 25 - Hierarchical Clustering/Hierarchical_Clustering/Mall_Customers.csv'
)
```

```python
x = in_df.iloc[:,[3,4]].values
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x, method='ward'))
plt.xlabel("Customer")
plt.ylabel("Distances")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
pred_hc = hc.fit_predict(x)
```

```python
# Visualization
plt.scatter(x[pred_hc==0,0], x[pred_hc==0,1], s=100, c='red', label='cluster 1')
plt.scatter(x[pred_hc==1,0], x[pred_hc==1,1], s=100, c='blue', label='cluster 2')
plt.scatter(x[pred_hc==2,0], x[pred_hc==2,1], s=100, c='green', label='cluster 3')
plt.scatter(x[pred_hc==3,0], x[pred_hc==3,1], s=100, c='cyan', label='cluster 4')
plt.scatter(x[pred_hc==4,0], x[pred_hc==4,1], s=100, c='magenta', label='cluster 5')

plt.xlabel("Annual Incom K$")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```
